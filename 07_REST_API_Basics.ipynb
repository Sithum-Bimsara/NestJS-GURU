## 1. What is a REST API?

* **REST** = **Re**presentational **S**tate **T**ransfer.
* It is a **software architectural style** that defines how **clients** (web browsers, mobile apps, other servers) interact with a **server** to **retrieve and manipulate data**.
* It is very popular because of:

  * **Simplicity**
  * **Scalability**
  * **Flexibility**
* **APIs can be of different types**.

  * An API is called a **REST API** only if it **follows the REST principles**.

## 3. REST Principle 1 â€“ Clientâ€“Server Architecture

* A REST API application always follows **clientâ€“server** architecture.
* There are two main parts:

  * **Client** â€“ sends requests (e.g., browser, mobile app, Postman).
  * **Server** â€“ receives the request, processes it, and sends back a response.
* Flow:

  1. **Client** sends an HTTP request to the **API** (hosted on a server).
  2. **Server/API** processes the request (validations, DB queries, etc.).
  3. **Server** returns an HTTP response back to the **client**.
* This separation allows the client UI and the server logic to evolve **independently**.

---

## 4. REST Principle 2 â€“ Stateless

* **Stateless** means:

  * Each HTTP request is **independent**.
  * The **server does not store any session state** between requests.
* Therefore:

  * **Every request must contain all the information** the server needs to understand and process it.

### Example â€“ Protected Resource (User Details)

* Suppose you want to call an endpoint to get user details: e.g., `/users/me`.

* You must be an **authenticated user**.

* In a stateless REST API:

  * The server **does NOT** store your login session like `userId` in memory.
  * Instead, **every request** to a protected resource must include something like an **authentication token** (e.g., JWT) in the request (usually in headers).

* Flow:

  1. Client logs in and receives an **auth token**.
  2. Client calls a protected endpoint like `/users/me` and sends the **token with the request**.
  3. Server reads and verifies the token:

     * If **valid** â†’ respond with user details.
     * If **invalid or expired** â†’ respond with an error (e.g., 401 Unauthorized).
  4. The server **doesnâ€™t store** this token or user session internally between requests.

> **Key idea:**
> No session data is stored on the server.
> All required info (like auth token) travels **with each request**.

## 5. REST Principle 3 â€“ Cacheability

* REST responses can be **cached** to improve performance.
* Scenario:

  * Client requests: `GET /users`
  * Server responds with a list of (say) **1000 users**.
  * Next time, the client makes the **same request**.
* Instead of hitting the server again and again for the same data:

  * The **client** or **intermediate caches** can store that response.
  * On repeated requests for the same resource, the data can be served from **cache**.

**Benefits:**

* Reduces **server load**.
* Improves **response time** and overall **user experience**.

> REST APIs should be **designed to allow caching** where appropriate (using correct cache headers, etc.).

## 6. REST Principle 4 â€“ Layered System

* REST allows a **layered architecture**.
* The client **does not know** how many layers the request passes through.

### Possible Layers on the Server Side

* **API Gateway / Reverse Proxy**

* **Middlewares** (authentication, logging, validation)

* **Cache layer** (e.g., Redis)

* **Application logic**

* **Database**

* When the client sends a request:

  * It may pass through multiple layers before reaching the final handler.
  * The client still sees a **single uniform API endpoint**.
  * The client does not care **how many internal layers** process the request.

> **Key idea:** The system is organized in layers, but this complexity is **hidden from the client**.

## 7. REST Principle 5 â€“ Uniform Interface

* REST uses a **uniform interface** between client and server.
* It relies on **standard HTTP methods (verbs)** to perform operations on **resources**.

### 7.1 Resources & Endpoints

* A **resource**: A thing or collection, e.g. `users`, `products`, `orders`.
* A **URL/endpoint** identifies the resource, e.g.:

  * `GET http://localhost:3000/users`

A request has **two main parts**:

1. **HTTP Method (Verb)** â€“ what action we want to perform.
2. **Endpoint (URL)** â€“ which resource weâ€™re targeting.

### 7.2 Same Endpoint, Different Methods

Example resource: `/users`

* **Get all users**:

  * `GET http://localhost:3000/users`
* **Create a new user**:

  * `POST http://localhost:3000/users`
* **Delete all users** (not common but possible):

  * `DELETE http://localhost:3000/users`

> Notice: The **URL is the same**.
> Only the **HTTP method changes**, and therefore the **action** changes.
> This is what **uniform interface** means.

## 8. Common HTTP Methods (Verbs)

### 8.1 GET

* **Purpose**: Retrieve data from the server.
* Examples:

  * Get all users â†’ `GET /users`
  * Get a single user by ID â†’ `GET /users/{id}`
  * Get all products â†’ `GET /products`

### 8.2 POST

* **Purpose**: Create a new resource on the server.
* Usage:

  * Send a **POST** request to an endpoint with a **request body**.
  * Body contains data needed to create the resource.
* Example:

  * `POST /products` with body `{ name, price, description }`.

### 8.3 PUT vs PATCH

* Both are used to **update existing resources**.

#### PUT

* **Full update**.
* You send the **entire object** in the request body, even if only one field changed.
* Example:

  * Update product â†’ `PUT /products/{id}`
  * Body must contain the whole product object.

#### PATCH

* **Partial update**.
* You send only the **fields you want to change**.
* Example:

  * Just change the `color` of a product â†’ `PATCH /products/{id}`
  * Body could be `{ "color": "blue" }` only.

### 8.4 DELETE

* **Purpose**: Delete a resource on the server.
* Example:

  * Delete product with ID 5 â†’ `DELETE /products/5`

> These five (GET, POST, PUT, PATCH, DELETE) are the **most commonly used** HTTP verbs in REST APIs.

## 9. REST Principle 6 â€“ Code on Demand (Optional)

* This is an **optional** REST constraint.
* It allows the server to send **executable code** to the client which the client can run.

  * Example: Downloading a script or some logic dynamically.
* This can extend client functionality **without a full app update**.
* In practice, this is **not used often** in typical API design, so we usually **ignore it** in backend courses.

## 10. Anatomy of an HTTP Request (Summary)

Whenever we call an API, the request generally contains:

1. **HTTP Method** (GET, POST, PUT, PATCH, DELETE, â€¦)

2. **Endpoint (URL)**

   * **Domain**

     * e.g. `http://localhost:3000`, `https://google.com`, `https://facebook.com`
   * **Route / Resource**

     * e.g. `/customers`, `/users`, `/posts`, `/products`
   * **Route Parameters (Path Parameters)** â€“ dynamic parts in the path

     * Example: `/customers/india`

       * `customers` â†’ route
       * `india` â†’ **route parameter** (can also be `usa`, `uk`, `sri-lanka`, etc.)
   * **Query Strings (Query Parameters)** â€“ extra info added after `?`

     * Example: `/customers/india?page=2&limit=10`
     * Used for filtering, pagination, sorting, etc.

3. **Request Body** (only for some methods)

   * Used mainly with **POST**, **PUT**, and **PATCH**.
   * Contains the actual **data** to create or update a resource.
   * Examples:

     * Creating a user â†’ `POST /users` with body `{ name, email, password }`.
     * Updating a product â†’ `PATCH /products/10` with body `{ price: 999 }`.
   * **GET** and **DELETE** usually **do not** have a request body.
